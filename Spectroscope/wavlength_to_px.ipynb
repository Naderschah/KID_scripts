### For the known equivalent wavlengths we can write:

$d\sin\theta_1=d\sin\theta_2/2=\lambda$

Which we can solve for the thetas, and then plug into the tan equation centered somewhere in the spectrum

which gives:

$y = x \frac{\lambda/d}{\sqrt{1-(\lambda/d)^2}}$

for first order, and for second we have each lambda term multiplied by 2
